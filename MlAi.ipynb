```python
print("hello")
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
area = df.area
area
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict(pd.DataFrame([[3000]], columns=['area']))
```

```python
area_df = pd.read_csv('areas.csv')
area_df
```

```python
area_df.head(4)
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices'] = p
area_df
```

```python
area_df.to_csv('prices.csv', index=False)
```

```python
# Keep the original hiring dataframe
hiring_df = pd.read_csv('hiring.csv')
hiring_df['experience'] = hiring_df['experience'].fillna("zero")
hiring_df
```

```python
import math
median_test_score = math.floor(hiring_df['test_score(out of 10)'].mean())
median_test_score
```

```python
hiring_df['test_score(out of 10)'] = hiring_df['test_score(out of 10)'].fillna(median_test_score)
hiring_df
```

```python
from word2number import w2n
hiring_df['experience'] = hiring_df['experience'].apply(w2n.word_to_num)
hiring_df
```
